```python
temperature_all = data['온도'].round().value_counts().sort_index()
temperature_all_list = temperature_all.index


for i in temperature_all_list:
    if not i in list(temperature['temperature']):

        temperature =temperature.append({'temperature' : i, 'counts' : 1},ignore_index=True)

temperature = temperature.sort_values('temperature')
temperature = temperature.reset_index(drop=True)

# temperature
    
# temperature_ratio.plot(kind='bar',x='temperature',y='counts')
# print("temp : ",row['온도']," counts :",temperature.loc[temperature['temperature'] == round(row['온도']), 'counts'].values[0])
```

```python
for idx, row in temperature.iterrows():
    temperature.at[idx,'temperature_ratio'] = temperature_all[row['temperature']] / row['counts']
```
