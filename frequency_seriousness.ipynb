```python
import platform
import os
import pandas as pd
import numpy as np
import sys
import ydata_profiling
import matplotlib.pyplot as plt
import math
from datetime import datetime

filename = '20221118_cid_nodrop.csv'
filepath = os.path.abspath(os.path.join(os.getcwd(),r'Results',filename))
plt.rc('font', family='Malgun Gothic')
```

```python
data = pd.read_csv(filepath,index_col=0,encoding='cp949')
```

```python
data['사고신고사유'].value_counts()
```

```python
for idx, row in data.iterrows():
    if '사망' in row['사고신고사유'] : 
        data.at[idx, 'seriousness'] = 4
    elif ('3일' in row['사고신고사유']) or ('1000만원' in row['사고신고사유']) :
        data.at[idx, 'seriousness'] = 3
    else:
        if row['사망자'] != 0 :
            data.at[idx, 'seriousness'] = 4
        else:
            damage = row['피해금액']
            if damage == "피해없음":
                data.at[idx, 'seriousness'] = 1
            elif damage == "1,000만원 미만":
                data.at[idx, 'seriousness'] = 2
            else:
                if "억원" in damage:
                    data.at[idx, 'seriousness'] = 4
                elif "만원" in damage:
                    data.at[idx, 'seriousness'] = 3
                else:
                    print(damage)
        

data['seriousness'].value_counts()
```

```python
ex_date = {}
dateformat = "%Y-%m-%d %H:%M"

# for idx, row in data.iterrows():
#     data.at[idx, '사고일시'] = row['사고일시'].strip()

for idx, row in data.sort_values('사고일시').iterrows():
    if row["작업프로세스"] in ex_date.keys():
        daydiff = (datetime.strptime(row["사고일시"], dateformat) - datetime.strptime(ex_date[row["작업프로세스"]], dateformat)).days
        print(row["작업프로세스"], " ",row["사고일시"], " " ,ex_date[row["작업프로세스"]], " ",  daydiff)
        if daydiff < 10:
            data.at[idx, 'frequency'] = 4
        elif daydiff < 20:
            data.at[idx, 'frequency'] = 3
        elif daydiff < 30:
            data.at[idx, 'frequency'] = 2
        else:
            data.at[idx, 'frequency'] = 1
        ex_date[row["작업프로세스"]] = row["사고일시"]
    
    else:
        ex_date[row["작업프로세스"]] = row["사고일시"]
        data.at[idx, 'frequency'] = 1
    # print(idx, " ", row["사고일시"])


data['frequency'].value_counts()
```
