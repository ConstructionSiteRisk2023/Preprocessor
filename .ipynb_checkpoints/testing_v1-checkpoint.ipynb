```python
def transform_working_type(x):
    if x == '콘크리트공사':
        return 50.33
    elif x == '기타공사':
        return 17.62
    elif x == '가설공사' :
        return 24.58
    elif x == '설비공사' : #설비공사 : 건물을 지을 때 생활에 밀접한 시설을 설치하는 필수 공사로 기초공사에서 예상한 구조를 완성시키는 과정
        return 34.04
    elif x == '토공사' :
        return 18.62
    elif x == '철골공사' :
        return 38.11
    elif x == '해체 및 재활용공사' :
        return 20.09
    elif x == '토목공사' : #일단 콘크리트로 ... 이거 어케해요ㅠ
        return 50.33
    elif x == '관공사' :
        return 17.62 # 기타로...
    elif x == '수장공사' :
        return 6.71
    elif x == '타일 및 테라코타공사' :
        return 18.18
    elif x == '미장공사' :
        return 32.36
    elif x == '내외벽공사' :
        return 32.83
    elif x == '도장공사' :
        return 18.90
    elif x == '목공사' :
        return 4.21
    elif x == '조적공사' :
        return 14.79
    elif x == '지정 및 기초공사' :
        return 34.04
    elif x == '금속공사' :
        return 8.54
    elif x == '방수 및 방습공사' :
        return 21.11
    elif x == '지붕 및 홈통공사' :
        return 85.64
    
```

```python
temp['공종'] = temp['공종'].apply(transform_working_type)
```

```python
def transform_temperature(x):
    if x < 0:
        return 0.196
    elif x>=0 and x < 4:
        return 0.163
    elif x>=4 and x<8:
        return 0.109
    elif x>=8 and x <12:
        return 0.082
    elif x>=12 and x<16:
        return 0.054
    elif x>=16 and x<20:
        return 0.043
    elif x >=20 and x<24:
        return 0.101
    elif x>=24 and x<28:
        return 0.125
    elif x>= 28:
        return 0.128
```

```python
temp['온도'] = temp['온도'].apply(transform_temperature)
```

```python
temp['공정율'].value_counts()
```

```python
def transform_working_per(x):
    if x =='90% 이상':
        return 0.1
    elif x=='20~29%':
        return 0.124
    elif x=='10~19%':
        return 0.092
    elif x=='10% 미만':
        return 0.083
    elif x=='30~39%':
        return 0.111
    elif x=='40~49%':
        return 0.085
    elif x=='50~59%':
        return 0.099
    elif x=='60~69%':
        return 0.111
    elif x=='70~79%':
        return 0.088
    elif x=='80~89%':
        return 0.107
    
```

```python
temp['공정율'] = temp['공정율'].apply(transform_working_per)
```

```python
def transform_month(x):
    if x==1:
        return 0.068
    elif x==2:
        return 0.086
    elif x==3:
        return 0.173
    elif x==4:
        return 0.068
    elif x==5:
        return 0.08
    elif x==6:
        return 0.086
    elif x==7:
        return 0.080
    elif x==8:
        return 0.111
    elif x==9:
        return 0.049
    elif x==10:
        return 0.086
    elif x==11:
        return 0.056
    elif x==12:
        return 0.056
```

```python
temp['Month'] = temp['Month'].apply(transform_month)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

scaler.fit(test)
test_scaled = scaler.transform(test)
test_scaled = pd.DataFrame(data = test_scaled)
```

```python
test_scaled
```

```python
from sklearn import model_selection
from sklearn.metrics import accuracy_score

import joblib

import warnings
warnings.filterwarnings("ignore")
```

```python
X_data = test_scaled
Y_data = np.array(y_data)

x_train, x_test, y_train, y_test = model_selection.train_test_split(X_data, Y_data, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import cross_val_score
```

```python
models = [
    ('Logistic', LogisticRegression()),
    ('KNN', KNeighborsClassifier(16)),
    ('SVM', SVC(kernel='rbf', C=100, gamma=0.01)),
    ('DecisionTree', DecisionTreeClassifier(random_state = 0, max_depth = 4)),
    ('RandomForest', RandomForestClassifier(random_state = 0, n_estimators = 100, max_depth = 6)),
    ('NaiveBayes', BernoulliNB(alpha=10))
]
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn import metrics
```

```python
for name, model in models:
    model.fit(x_train, y_train)
    print(f'{name} _accuracy: ', metrics.accuracy_score(model.predict(x_test), y_test))
    print(f'{name} _precision: ', metrics.precision_score(model.predict(x_test), y_test,average='micro'))
    print(f'{name} _recall: ', metrics.recall_score(model.predict(x_test), y_test,average='micro'))
    print(f'{name} _f1 score: ', metrics.f1_score(model.predict(x_test), y_test,average='micro'))
    print(" ")
```
