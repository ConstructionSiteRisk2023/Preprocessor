```python
import platform
import os
import pandas as pd
import numpy as np
import sys

filename = 'ta_20230425145448.csv'
filepath = os.path.abspath(os.path.join(os.getcwd(),r'Data',filename))
filepath
```

```python
data = pd.read_csv(filepath)
```

```python
# for idx, row in data.iterrows():
#     data['average(cel)'] = 

# data['average(cel)']
data = data.round()
data['average(cel)'].value_counts()
average_counts = data['average(cel)'].value_counts(dropna=True, sort=True)
highest_counts = data['highest(cel)'].value_counts(dropna=True, sort=True)
lowest_counts = data['lowest(cel)'].value_counts(dropna=True, sort=True)
```

```python
# type(value_counts)
# indexlist = list(value_counts.index)
# indexlist.sort()
# indexlist


indexlist = average_counts.index

for i in range(-14,37):
    if not i in indexlist:
        average_counts.loc[i] = 0
average_counts = average_counts.sort_index()
highest_counts = highest_counts.sort_index()
lowest_counts = lowest_counts.sort_index()

# average_counts

average_counts = average_counts.to_frame()
highest_counts = highest_counts.to_frame()
lowest_counts = lowest_counts.to_frame()

average_counts= average_counts.rename(columns={"average(cel)" : "temperature"})
highest_counts= highest_counts.rename(columns={"highest(cel)" : "temperature"})
lowest_counts= lowest_counts.rename(columns={"lowest(cel)" : "temperature"})
average_counts
```

```python


for idx, row in highest_counts.iterrows():
    average_counts.at[idx,'temperature'] += row

for idx, row in lowest_counts.iterrows():
    average_counts.at[idx,'temperature'] += row
average_counts
```

```python
for idx, row in average_counts.iterrows():
    if row['temperature'] == 0:
        average_counts.at[idx,'temperature'] = 1

average_counts
```

```python
average_counts.to_csv(os.path.abspath(os.path.join(os.getcwd(),r'Results','temperature_dist.csv')))
```
